## 1. Add Unit Tests
Unit tests are essential for ensuring code reliability and correctness. Below is an example of how to write and execute unit tests using Python's `unittest` module.

```python
import unittest
def add(a, b):
    return a + b

class TestAddition(unittest.TestCase):
    def test_add_positive_numbers(self):
        self.assertEqual(add(2, 3), 5)
    def test_add_negative_numbers(self):
        self.assertEqual(add(-2, -3), -5)

if __name__ == "__main__":
    unittest.main()
```

## 2. Optimize Code
Optimizing code involves improving its efficiency and readability. Below is an example of optimizing a function to calculate the sum of squares using Python's built-in functions.

```python
# Original function
def sum_of_squares(numbers):
    result = 0
    for n in numbers:
        result += n ** 2
    return result

# Optimized function using list comprehension
def optimized_sum_of_squares(numbers):
    return sum(n ** 2 for n in numbers)

# Example usage
numbers = [1, 2, 3, 4]
print("Original:", sum_of_squares(numbers))
print("Optimized:", optimized_sum_of_squares(numbers))
```

## 3. Jupyter Notebook vs Python Script for Results and Analysis
Jupyter notebooks are widely used in the AI/ML industry for interactive exploration and visualization. However, Python scripts are preferred for production and reproducibility. Below is a comparison:
- **Jupyter Notebooks:** Ideal for prototyping, visualization, and sharing insights interactively.
- **Python Scripts:** Better for automation, version control, and integration into larger systems.
### Industry Practices
- Use Jupyter notebooks for initial exploration and presentations.
- Transition to Python scripts for production workflows.
